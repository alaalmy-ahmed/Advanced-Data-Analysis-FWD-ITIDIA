```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_total = pd.read_csv('countries_total.csv')
sns.barplot(data=customer_total, y='Country', x='Countries_totals')
plt.title('Countries total Invoices')
plt.show()
```

```python
# Which customer from the USA purchased the most?

usa_customers = pd.read_csv('usa_customers.csv')
usa_customers['full name'] = usa_customers['FirstName'] + ' ' + usa_customers['LastName']
plt.figure(figsize=(20, 10))
sns.barplot(data=usa_customers, x='full name', y='Total_purchase')
plt.title('top 10 Invoice totals per Customer in usa')
plt.ylabel('Totals')
plt.show()
```

```python
no_albumns = pd.read_csv('top_10_artists.csv')
sns.barplot(data=no_albumns, y='Artist_name', x='NO_albums')
plt.title('top 10 artists producing albums')
plt.show()
```

```python
# Which artist made the most sales?

artists = pd.read_csv('top 10 artists total.csv')
plt.figure(figsize=(20, 8))
sns.barplot(data=artists, x='Name', y='Total')
plt.title('top 10 artists purchases')
plt.ylabel('Totals')
plt.show()
```
